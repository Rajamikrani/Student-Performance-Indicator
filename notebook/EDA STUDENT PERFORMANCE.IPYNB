- Adding Column for Total_Score and Average_Score

```python
df['Total_Score'] = df['math_score'] + df["reading_score"] + df['writing_score']
df['Average_Score'] = df['Total_Score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['Total_Score'].count()
writing_full = df[df['writing_score'] == 100]['Total_Score'].count()
math_full = df[df['math_score'] == 100]['Total_Score'].count()
# ['Total_Score'].count() this is no used for counting the filtered rows
# we can use any features instead of it

print("Number of student Scores full marks in Math : " , math_full)
print("Number of student Scores full marks in Writing : " , writing_full)
print("Number of student Scores full marks in Reading : " , reading_full)
```

```python
reading_less_20 = df[df['reading_score'] < 20]['Total_Score'].count()
writing_less_20 = df[df['writing_score'] < 20]['Total_Score'].count()
math_less_20 = df[df['math_score'] < 20]['Total_Score'].count()
# ['Total_Score'].count() this is no used for counting the filtered rows
# we can use any features instead of it

print("Number of student with less than 20 marks in Math : " , math_less_20 )
print("Number of student with less than 20 marks in Writing : " , writing_less_20 )
print("Number of student with less than 20 marks in Reading : " , reading_less_20 )
```

### Insights
- From above values we get students have performed the worst in Maths
- Best performance is in reading section

### BIVARIATE ANALYSIS ( Is gender has any impact on student's performance ? )

```python
gender_group = df.groupby('gender').mean(numeric_only=True)
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']

female_scores = [gender_group['Average_Score'][0], gender_group['math_score'][0]]
male_scores = [gender_group['Average_Score'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
```

#### Insights
- On an average females have a better overall score than men.
- whereas males have scored higher in Maths.
